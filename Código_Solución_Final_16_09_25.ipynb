```python
#Juntando filtro de item + descripcion en uno solo

Algoritmo_busqueda=data[data["ITEM"].str.contains(A)|data["DESCRIPCION"].str.contains(A)]
```

```python
data.head() ##
```

```python
# identificando los proveedores que son true o false

proveedor_bool=[]
for i in data.columns:
  if data[i].dtype=="bool":
    proveedor_bool.append(i)

```

```python
proveedor_bool
```

```python
# Algortimo de coincidencia para la busqueda y proveedor

resultados=[]
```

```python
for _, i in Algoritmo_busqueda.iterrows():
  proveedores_disponibles=[]
  # recorriendo los proveedores y validando si son true
  for j in proveedor_bool:
    if i[j]==True:
      proveedores_disponibles.append(j)
  resultados.append({
      "ITEM": i["ITEM"],
      "DESCRIPCION": i["DESCRIPCION"],
      "PROVEEDORES": proveedores_disponibles
  })
```
